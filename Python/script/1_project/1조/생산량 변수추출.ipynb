```python
import numpy
from sklearn.model_selection import train_test_split
from sklearn import ensemble
from sklearn.metrics import mean_squared_error
import pylab as plot


X_train, X_test, y_train, y_test = train_test_split(X , y, random_state=0)

#randomF= RandomForestRegressor(n_estimators=100,max_depth =3,min_samples_leaf=1).fit(X_train,y_train)
#print('정확도:%.2f'%(randomF.score(scaled_data,y)*100),"%")


mseOos = []
nTreeList = range(50, 500, 10)
for iTrees in nTreeList:
    depth = None
    maxFeat = 3 #조정해볼 것
    randomF = ensemble.RandomForestRegressor(n_estimators=iTrees,
                    max_depth=depth, max_features=maxFeat,
                    oob_score=False, random_state=531)
    randomF.fit(X_train, y_train)
    
    #데이터 세트에 대한 MSE 누적
    prediction =randomF.predict(X_test)
    mseOos.append(mean_squared_error(y_test, prediction))
    
  
print("MSE")
print(mseOos[-1])

# 트레이닝 테스트 오차에 대한 앙상블의 트리개수 도표그리기


plot.plot(nTreeList, mseOos)
plot.xlabel('Number of Trees in Ensemble')
plot.ylabel('Mean Squared Error')
```

```python
# 우리 독립변수들간의 중요도 도표그리기
featureImportance = randomF.feature_importances_

# 가장 중요도 높은 기준으로 ..
featureImportance = featureImportance/featureImportance.max()
sorted_idx = numpy.argsort(featureImportance)
barPos = numpy.arange(sorted_idx.shape[0])+.5
plot.barh(barPos, featureImportance[sorted_idx], align='center')
plot.yticks(barPos, Xnames[sorted_idx])
plot.xlabel('Variable Importance')
plot.show()
```
